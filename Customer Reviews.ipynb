```python
text1= ' '.join(map(str, dfGO1.Body.values)) 
  
print(text1)  
```

```python
text2= ' '.join(map(str, df1N1.Body.values)) 
  
print(text2)  
```

### Outdoor Go1

```python
import re
text1 = re.sub(r'==.*?==+', '', text1)
text1 = text1.replace('\n', '')

text2 = re.sub(r'==.*?==+', '', text2)
text2 = text2.replace('\n', '')

def plot_cloud(wordcloud):
    plt.figure(figsize=(40, 30))
    plt.imshow(wordcloud) 
    plt.axis("off");
    
from wordcloud import WordCloud, STOPWORDS

wordcloud = WordCloud(width = 1500, height = 500, random_state=1, background_color='salmon', 
                      colormap='Pastel1', collocations=False, stopwords = STOPWORDS).generate(text1)

plot_cloud(wordcloud)
```

For outdoor Products, users care more about: 
- `motion`,`app`,`battery`

### Indoor 1N1

```python
wordcloud = WordCloud(width = 1500, height = 500, random_state=1, background_color='salmon', 
                      colormap='Pastel1', collocations=False, stopwords = STOPWORDS).generate(text2)

plot_cloud(wordcloud)
```

For indoor Products, users care more about:
