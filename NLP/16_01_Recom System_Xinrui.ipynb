#### B) Multiprocessing

```python
# Multiprocessing vs threading, queue
```

```python
users='steam_user_id.txt'
with open(users,'r') as f:
    user_id=f.readlines()[:50]
```

```python
def worker(user_id_temp):
    dic_temp={}
    for user_id in tqdm(user_id_temp,leave=False):
        url="http://api.steampowered.com/IPlayerService/GetOwnedGames/v0001/"
        params={
        'key':'EB0DBACF35F00F57AF491F955CD9A03A',
        'steamid':user_id.strip(),
        'format':'json'
    
        }
    r=requests.get(url, params=params)
    user_inventory=r.json().get('response').get('games')
    
    dic_temp.update({user_id.strip():user_inventory})
    
    time.sleep(.5)
    return dic_temp
```

```python
p=Pool(2)

dic_master={}
for i in tqdm(list(split_list(user_id,10))):
    temp_dic=p.map(worker, split_list(i,5))
    for j in temp_dic:
        dic_master.update(j)
    
    time.sleep(.5)
```
